# 1- Import

```python
""" 
Importing packages for task execution
"""

from datetime import date, datetime, timedelta
from typing import Optional
from groq import Groq
```

# 2- Constants

```python
"""
Sets up the execution environment loading the env, stopping the loop and creating variables
"""

GROQ_CLIENT = None
```

# 3- Pre Processing

```python
# Configure the Groq Client in main

def set_client(client) -> None:
    global GROQ_CLIENT
    GROQ_CLIENT = client
```

# 4- Processing

```python

"""
    Setting the local context using a dictionary ang getting a text generated by Groq with 25 words
"""


# Gerenerate a Synopsis about a specific movie
def generate_synopsis(title: str, max_words: int = 60, style_hint: Optional[str] = None) -> str:
    if GROQ_CLIENT is None:
        raise RuntimeError("GROQ_CLIENT not configured. Use set_client().")

    try:
        prompt = f"Write a concise synopsis for the movie titled: '{title}'. Limit to {max_words} words."
        if style_hint:
            prompt += f" Style: {style_hint}."

        chat_completion = GROQ_CLIENT.chat.completions.create(
            messages=[{"role": "user", "content": prompt}],
            model="llama-3.1-8b-instant",
        )

        synopsis = chat_completion.choices[0].message.content.strip()
        words = synopsis.split()
        if len(words) > max_words:
            synopsis = " ".join(words[:max_words])

        return synopsis
    except Exception as e:
        raise RuntimeError(f"Error to generete a synopsis with groq: {e}")
```
